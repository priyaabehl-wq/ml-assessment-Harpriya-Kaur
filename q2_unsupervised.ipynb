```python
import pandas as pd

df = pd.read_csv('q2_customers.csv')
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

### Feature Scaling

Feature scaling is essential before applying K-Means clustering because:

- K-Means uses distance-based calculations (Euclidean distance)
- Features with larger values (e.g., annual_spend) would dominate smaller ones (e.g., visits_per_month)
- Scaling ensures all features contribute equally to cluster formation

Therefore, StandardScaler is used to standardize the data.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

### Elbow Method

The elbow plot shows a sharp decrease in WCSS from K = 1 to K = 3.

After K = 3, the rate of decrease slows down significantly, indicating diminishing returns from adding more clusters.

Therefore, the optimal number of clusters is chosen as K = 3.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

df['cluster'] = clusters
```

```python
df.head()
```

```python
centroids = pd.DataFrame(
    scaler.inverse_transform(kmeans.cluster_centers_),
    columns=df.columns[:-1]
)

centroids
```

### Cluster Interpretation

Each cluster represents a group of customers with similar behavior:

- Cluster 0: Moderate spending and average visits
- Cluster 1: High spenders with frequent visits
- Cluster 2: Low spenders with infrequent visits

These segments can help businesses target customers with personalized strategies.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
```

```python
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
loadings = pd.DataFrame(
    pca.components_,
    columns=df.columns[:-1],
    index=['PC1', 'PC2']
)

loadings
```

### PCA Interpretation

- PC1 captures the most variance in the dataset
- PC2 captures the second most variance

Based on loadings:
- PC1 is influenced by features with high values in the first component
- PC2 captures variation from other features

This helps reduce dimensionality while retaining important information.

```python
plt.scatter(pca_data[:,0], pca_data[:,1], c=clusters)
plt.title('Customer Segmentation (PCA)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

### Cluster Visualization

- The scatter plot shows customer segments in 2D space
- Each color represents a cluster
- Well-separated clusters indicate good segmentation
